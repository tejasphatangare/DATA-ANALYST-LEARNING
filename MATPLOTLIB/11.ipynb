# SAVEFIG-it is used for save the figures in diff format

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,2,1,5,4]

plt.plot(x,y)
plt.savefig('Linear')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,1,5,4]

plt.plot(x,y)
plt.fill_between(x,y,color='pink')
plt.savefig('Linear')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,1,5,4]

plt.plot(x,y)
plt.fill_between(x,y,color='pink')
plt.savefig('Myfile1.pdf',dpi=2000)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,1,5,4]

plt.plot(x,y)
plt.fill_between(x,y,color='pink')
plt.savefig('Myfile2.jpg',dpi=2000,facecolor='Green')
plt.show()
```
