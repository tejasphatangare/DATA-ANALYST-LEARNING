# STEAM PLOT-

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5]
Hours_Worked=[2,4,1,3,5]

plt.stem(Day,Hours_Worked)
plt.show()
```

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5]
Hours_Worked=[2,4,1,3,5]

plt.stem(Day,Hours_Worked,linefmt=':',markerfmt='*',bottom=1,basefmt='Red',label='Hours_Worked',use_line_collection=False)

plt.legend()
plt.show()


#ERROR DUE TO FUNCTION WILL REMOVE LATER
```
