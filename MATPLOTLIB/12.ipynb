# xticks- it is used for set the values or lables of x axis-

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,1,2,3,5]

plt.plot(x,y)
plt.xticks(x)
plt.show()
```

# yticks-it is used for set the value of y axis-

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,1,2,3,5]

plt.plot(x,y)
plt.yticks(x)
plt.show()
```

# xlim-it is used for increse the value of an x axis.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,1,2,3,5]

plt.plot(x,y)
plt.xlim(0,20)
plt.show()
```

# ylim-it is used for increse the value of an y axis -

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,1,2,3,5]

plt.plot(x,y)
plt.xlim(0,20)
plt.ylim(0,10)
plt.show()
```
