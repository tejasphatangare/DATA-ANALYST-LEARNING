```python
import matplotlib.pyplot as plt
```

# Text in Plot-

```python
x=[1,2,3,4,5]
y=[3,5,1,2,4]

plt.plot(x,y)
plt.text(2,5,'PEAK_VALUE',fontsize=10,style='italic')
plt.show()
```

# Annoate Function in  plot-

```python
x=[1,2,3,4,5]
y=[3,5,1,2,4]

plt.plot(x,y)
plt.annotate('Peak Watching',xy=(2,5),xytext=(3,5),arrowprops=dict(facecolor='Black',shrink=100))
plt.show()
```

# legend in matplotlib-

```python
x=[1,2,3,4,5]
y=[3,5,1,2,4]

plt.plot(x,y)

plt.legend(['Watching'],loc=9)
plt.show()
```
