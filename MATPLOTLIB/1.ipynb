# BAR GRAPH-

```python
import matplotlib.pyplot as plt
```

```python
st='Om','Sai','Ram','Mayuri','Pooja'
mk=75,80,85,92,97


plt.bar(st,mk)
plt.show()
```

```python
st='Om','Sai','Ram','Mayuri','Pooja'
mk=97,60,45,92,81

plt.xlabel('Students',fontsize=10)
plt.ylabel('Marks',fontsize=10)
plt.title('Unit-Test-1')


col=['red', 'blue', 'green', 'orange','yellow']
plt.bar(st,mk,width=0.5,color=col,edgecolor='black',linewidth=0.5,label=90)
plt.legend()
plt.show()
```

# MULTIPLE BAR GRAPH IN SINGLE GRAPH-

```python
import numpy as np

st=['pooja','sai','priya','om','divya','ram']
eng=[96,87,53,89,78,95]
math=[44,55,88,77,67,97]

width=0.4
x1=np.arange(len(st))
x2=[j+width for j in x1]


plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Test_Marks')

plt.bar(x1,eng,width,edgecolor='Indigo',linewidth=0.5,label='English')
plt.bar(x2,math,width,edgecolor='Black',linewidth=0.5,label='Math')

plt.xticks(x1+width/2,st,rotation=20)
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np

st=['pooja','sai','priya','om','divya','ram']
eng=[96,87,53,89,78,95]
math=[44,55,88,77,67,97]

width=0.4
x1=np.arange(len(st))
x2=[j+width for j in x1]


plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Test_Marks')

plt.barh(x1,eng,width,edgecolor='Indigo',linewidth=0.5,label='English')
plt.barh(x2,math,width,edgecolor='Black',linewidth=0.5,label='Math')

plt.xticks(x1+width/2,st,rotation=20)
plt.legend(loc='best')
plt.show()
```
