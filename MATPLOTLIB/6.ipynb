# STACK PLOT-

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area1=[5,4,1,2,3]



l=['area1',]
c=['indigo']

plt.stackplot(x,area1,labels=l,colors=c)


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=1)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[5,4,3,2,1]
area2=[3,1,2,5,4]
area3=[1,2,3,4,5]


l=['area1','area2','area3']
c=['red','indigo','yellow']

plt.stackplot(x,area1,area2,area3,labels=l,colors=c)


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=2)
plt.show()
```
