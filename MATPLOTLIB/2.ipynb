```python
import matplotlib.pyplot as plt
```

```python
num1=[8,9,2,3,1,6,5]
num2=[2,3,1,7,5,9,4]
day=[1,2,3,4,5,6,7]

plt.scatter(num1,day)
plt.show()
```

```python
num1=[8,9,2,3,1,6,5]
num2=[2,3,1,7,5,9,4]
day=[1,2,3,4,5,6,7]

# VARIABLES START
colors=['Red','Green','Blue','yellow','orange','pink','Indigo']
sizes=[10,20,30,40,50,60,70]
# VARIABLES END


plt.scatter(num1,day,color=colors,s=sizes,marker='*',edgecolor='Red',linewidth=0.5,label='No Of Days Attendence')
plt.title('Scatter Plot',fontsize=10)
plt.xlabel('Days')
plt.ylabel('Numbers')
plt.legend(loc='best')
plt.show()
```

# MULTIPLE SCATTERS IN SINGLE SCATTER PLOT-

```python
num1=[8,9,2,3,1,6,5]
num2=[2,3,1,7,5,9,4]
day=[1,2,3,4,5,6,7]



plt.scatter(num1,day,color='black',marker='*',edgecolor='Black',linewidth=0.7,label='Week1',s=24)
plt.scatter(num2,day,color='Green',marker='*',edgecolor='Red',linewidth=0.5,label='Week-2',s=26)
plt.title('Scatter Plot',fontsize=10)
plt.xlabel('Days')
plt.ylabel('Numbers')
plt.legend(loc='best')
plt.show()
```
