# fill_between plot-

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[3,1,2,5,4]

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,2,5,4]

plt.plot(x,y)
plt.fill_between(x,y,color='orange',alpha=0.3)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,1,2,5,4])

plt.plot(x,y)
plt.fill_between(x,y,color='orange',alpha=0.3,where=(x>=2) & (x<=4))
plt.show()
```
