# STEP PLOT-

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,1,4,3,5]

plt.step(x,y,marker='*',color='Red',markersize=10,markerfacecolor='Green')

plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,4,3,5]

plt.step(x,y,marker='*',color='Orange',markersize=10,markerfacecolor='Black',label='Positivity')

plt.title('Step Plot',fontsize=10)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,4,3,5]
y1=[5,3,4,2,1]


plt.step(x,y,marker='*',color='indigo',markersize=10,markerfacecolor='Black',label='+ve')
plt.step(x,y1,marker='*',color='indigo',markersize=10,markerfacecolor='Black',label='-ve')
plt.title('Step Plot',fontsize=10)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=4)
plt.grid()
plt.show()
```
