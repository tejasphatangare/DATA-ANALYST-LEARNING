```python
data = [32, 96, 45, 67, 76, 28, 79, 62, 43, 81, 70,
61, 95, 44, 60, 69, 71, 23, 69, 54, 76, 67,
82, 97, 26, 34, 18, 16, 59, 88, 29, 30, 66,
23, 65, 72, 20, 78, 49, 73, 62, 87, 37, 68,
81, 80, 77, 92, 81, 52, 43, 68, 71, 86]

mybins=[20,30,40,50,60,70,80,90,100]

plt.hist(data,color='Orange',edgecolor='Black',bins=mybins,cumulative=-1)

plt.title('Histogram Plot',fontsize=10)
plt.xlabel('No Of Days',fontsize=10)
plt.ylabel('Values',fontsize=10)
plt.show()
```
