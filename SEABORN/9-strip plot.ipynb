```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tejas\Downloads\tips.csv")
df
```

```python
sns.stripplot(y='total_bill',x='time',data=df)
plt.show()
```

```python
sns.stripplot(y='total_bill',x='time',data=df,hue='day')
plt.show()
```

```python
sns.stripplot(y='total_bill',x='time',data=df,hue='sex',linewidth=0.7,edgecolor='red',palette='bright')
plt.show()
```

```python
sns.stripplot(y='total_bill',x='time',data=df,hue='sex',jitter=0.1)
plt.show()
```

```python
sns.stripplot(y='total_bill',x='time',data=df,hue='sex',alpha=0.5)
plt.show()
```

```python
sns.stripplot(x=df['total_bill'],jitter=0.1)
plt.show()
```
