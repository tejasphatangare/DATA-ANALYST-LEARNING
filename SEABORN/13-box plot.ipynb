```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tejas\Downloads\tips.csv")
df
```

```python
sns.boxplot(x='time',y='total_bill',data=df)
plt.show()
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex',color='gold')
plt.show()
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex',order=['Lunch','Dinner'])
plt.show()
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex',color='white',showmeans=True)
plt.show()
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex',color='White',
            showmeans=True,meanprops={'marker':'*','markeredgecolor':'red'})
plt.show()
```

```python
sns.boxplot(x='total_bill',y='time',data=df)
plt.show()
```

```python
sns.boxplot(x='total_bill',y='tip',data=df,orient='horizontal')
plt.show()
```
