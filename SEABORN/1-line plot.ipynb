```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LINEPLOT FROM USING DATAFRAME-
d1=[1,2,3,4,5]
d2=[3,5,4,1,2]

df=pd.DataFrame({'x':d1,'y':d2})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```
