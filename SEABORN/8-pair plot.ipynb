```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tejas\Downloads\tips.csv").head(100)
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue='sex')
plt.show()
```

```python
sns.pairplot(df,hue='sex',hue_order=['Male','Female'])
plt.show()
```

```python
sns.pairplot(df,hue='sex',hue_order=['Male','Female'],vars=['total_bill','tip'],palette='bright')
plt.show()
```

```python
sns.pairplot(df,hue='sex',palette='bright',y_vars='total_bill',x_vars='day')
plt.show()
```

```python
#kind-scatter,kde,hist,reg
sns.pairplot(df,hue='sex',palette='bright',kind='reg',markers=['*','<'])
plt.show()
```
