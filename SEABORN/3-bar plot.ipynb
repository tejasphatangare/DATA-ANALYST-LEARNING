```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tejas\Downloads\penguins.csv")
df
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',)
plt.show()
```

```python
ord=['Dream','Biscoe','Torgersen']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',order=ord,hue_order=['FEMALE','MALE'],ci=0)
plt.show()
```
